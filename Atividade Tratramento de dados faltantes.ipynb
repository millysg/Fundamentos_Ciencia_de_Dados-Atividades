# Por que você diria que isso acontece, no caso desses dados?

## Esses dados de Ttrab são o tipo de trabalho da pessoa, porém tem pessoas que no Trab responderam que não trabalham ou são aposentados, já o Itrab, que é a idade que começou a trabalhar, é possível que os participantes não quiseram mencionar essa informação. Já o Serief, tem a opção de deixar em branco se não parou de estudar.

```python
# Sustituindo dados faltantes pela média
media_Ttrab = dados['Ttrab']. mean()
media_Itrab = dados['Itrab']. mean()
media_Serief = dados['Serief']. mean()

dados['Ttrab'].fillna(media_Ttrab, inplace = True)
dados['Itrab'].fillna(media_Itrab, inplace = True)
dados['Serief'].fillna(media_Serief, inplace = True)


dados.head()
```
